```python
import matplotlib.cm as cm
from matplotlib.colors import Normalize
import matplotlib.pyplot as plt
import numpy as np


import pickle

with open("oct_spectator_scan.pickle", "rb") as data_file:
    oct_scan = pickle.load(data_file)

with open("drag_spectator_scan.pickle", "rb") as data_file:
    drag_scan = pickle.load(data_file)
```

DRAG scan

```python
drag_scan.keys()
```

```python
results = drag_scan["results"]
angles = drag_scan["angles"]
reps = drag_scan["reps"]

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,4),dpi=200)
cmap = cm.get_cmap('coolwarm')



titles = [r'Control (Q17) $|-,0\rangle$ prep', 'Spectator (Q18)']
for i,ax in enumerate(axes.flat):
    im = ax.imshow(np.flip(results[:,:,i],axis=0),extent=[angles[0]*2,angles[-1]*2,reps[0],reps[-1]],
                 aspect='auto',cmap='viridis',vmin=0,vmax=.2)
    ax.set_title(titles[i])
    if i == 0:
        ax.set_ylabel('N')
fig.colorbar(im, ax=axes.ravel().tolist(),extend='both',label='P(1)')


fig.supxlabel(r'$\phi$')

fig.suptitle(r'%.0f ns DRAG $X_{\pi/2}$'%(144/4.5),x=.43,y=1.0)
```
